### preprocessing

```python
def preprocess_text(text):
    text = text.lower()
    chars_to_remove = string.punctuation + string.digits + '«»' +'—'+'―'
    translator = str.maketrans('', '', chars_to_remove)
    text = text.translate(translator)
    tokens = text.split()
    return tokens

tokens = preprocess_text(text_data)
```

```python
tokens
```

### markov model

```python
def build_markov_model(tokens):
    model = {}
    for i in range(len(tokens) - 1):
        current_word = tokens[i]
        next_word = tokens[i+1]

        if current_word in model:
            model[current_word].append(next_word)
        else:
            model[current_word] = [next_word]

    return model

markov_model = build_markov_model(tokens)
```

### generate func

```python
def generate_poetry(model, length=70):
    if not model:
        return "Մոդելը դատարկ է։ Հնարավոր է՝ տվյալների ֆայլը դատարկ է կամ չի գտնվել։"

    start_word = random.choice(list(model.keys()))
    poetry = [start_word]
    current_word = start_word

    for _ in range(length - 1):
        if current_word not in model:
            current_word = random.choice(list(model.keys()))
            continue

        possible_next_words = model[current_word]
        next_word = random.choice(possible_next_words)
        poetry.append(next_word)
        current_word = next_word

    return ' '.join(poetry)
```

### text generation

```python
generated_text = generate_poetry(markov_model, length=70)

print(generated_text)
```
